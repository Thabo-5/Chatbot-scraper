```python
doc_dir = "./txt_files"
```

```python
for filename in os.listdir(doc_dir):
    if filename.endswith('.txt'):
        with open(os.path.join(doc_dir, filename)) as file:
            text = file.read().split('\n')
            for line in range(len(text)):
                if text[line].endswith("?") and text[line].startswith("H") ==False:
                    print(text[line])
                    for text[line+1] in text:
                        if text[line] != '':
                            print(line)
```
